### 1. 파이썬 언어에서는 숫자형(Numbers), 논리형(Bool), 시퀀스형(Sequence), 매핑형(Mapping) 집합형(Sets) 자료형 제공





### 파이썬 자료형

| **자료형**       | **자료형 예**    | **비 고**        |
|------------------|-----------------------|-----------------------|
| **int**          | 42                    | 소수점 없는 숫자   |
| **float**        | 3.14                  | 소수점 있는 숫자  |
| **complex**      | 2 + 3j                | 복소수 숫자       |
| **bool**         | True 또는 False       | 참, 거짓 값      |
| **str**          | "hello world" 또는 'hello world'      | 문자 순서 열(첨자로 참조 가능)       |
| **list**         | [1, 2, 3, 2]            | 변경이 가능한(mutable) 원소의 순서 열(첨자로 참조 가능)       |
| **tuple**        | (1, 2, 3)            | 변경이 불가능한(immutable) 원소의 순서 열(첨자로 참조 가능)       |
| **dict**         | {'k1':10, 'k2':20}   | 키(key)와 값(value)의 매핑이 필요한 경우, 순서 없음(첨자로 참조 불가) |
| **set**          | {1, 2, 3}            | 변경이 가능한(mutable) 유일한 원소들의 집합 , 순서 없음(첨자로 참조 불가)      |
| **frozenset**    | frozenset({1, 2, 3}) | 변경이 불가능한(immutable) 유일한 원소들의 집합 , 순서 없음(첨자로 참조 불가)      |
| **range**        | ranage(5)           | 변경이 불가능한(immutable) 숫자열 (0, 1, 2, 3, 4)     |




### 2. (LAB-1 : 따라하기) 파이썬에서 각 자료형의 변수 생성해 보기

### 2.1 숫자형 (int, float, complex)

```python
# int 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
x = 1        # x = int(1)             # 정수 값 1로 int 인스턴스 x 생성
print(x, type(x))
```

```python
# float 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
y = 2.8      # y = float(2.8)         # 실수 값 2.8로 float 인스턴스 y 생성
print(y, type(y))
```

```python
# compex 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
z = 1 + 3j   # z = complex(1 + 3j)    # 복소수 값 (1+3j)으로 complex 인스턴스 y 생성
print(z, type(z))
```

### 2.2 논리형(bool)

```python
# 논리형 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
a = True           # bool(True)       # True 값으로 bool 인스턴스 a 생성
b = False          # bool(False)      # False 값으로 bool 인스턴스 b 생성
c = 3 > 4          # bool(3 > 4)      # (3 > 4) 논리 연산 값으로 bool 인스턴스 c 생성
print(a, type(a))
print(b, type(b))
print(c, type(c))
```

```python
# 논리형 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
d = bool(0)        # 0, 0.0, Null 문자열 값으로 bool 인스턴스 생성 : False
e = bool(0.0)
f = bool('')
print(d, type(d))
print(e, type(e))
print(f, type(f))
```

```python
# 논리형 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
g = bool(2)        # 0, 0.0, Null 문자열이 아닌 값으로 bool 인스턴스 생성 : True
h = bool(1.0)
i = bool('a')
print(g, type(g))
print(h, type(h))
print(i, type(i))
```

### 2.3 시퀀스 자료형 - 문자열(str)
- 순서가 있음 : 첨자로 참조할 수 있음


```python
# 문자열 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
a = 'hello'    # str('hello')         # 'hello' 값으로 str 인스턴스 a 생성
b = "world"    # str("world")         # "world" 값으로 str 인스턴스 b 생성
print(a, type(a))
print(b, type(b))
```

```python
# 여러 줄의 문자열 값으로 str 인스턴스 c 생성
c = '''My name
is Jung'''
print(c, type(c))
```

```python
# 문자열 내부의 문자열 표현
d = "'python'"
print(d)
```

```python
# 문자열 내부의 문자열 표현
e = '"python"'
print(e)
```

```python
# 시퀀스형 자료 (순서를 가지고 나열된 형태)
f = 'dongyang'
print(f[0], f[1], f[2])
```

```python
# 시퀀스형 자료 (순서를 가지고 나열된 형태)
# 제어문을 활용하여 한 글자씩 출력하기
g = 'computer'
for char in g:
    print(char)
```

```python
# 파이썬에서는 한 글자의 문자도 문자열로 취급함
h = 'a'
print(h[0])
```

### 2.4 시퀀스 자료형 - 리스트(list)
- 변경이 가능한(mutable) 원소의 순서 열
- 순서가 있음 : 첨자로 참조할 수 있음


```python
# 리스트 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
a = [10, 20, 30]    # list([10, 20, 30])        # [10, 20, 30] 값으로 list 인스턴스 a 생성
print(a, type(a))
```

```python
# 리스트 자료 항목으로 다양한 자료형 사용 가능
b = [1, 3.92, 3 > 5, 'dongyang']    # list([1, 3.92, 3 > 5, 'dongyang'])        # [1, 3.92, 3 > 5, 'dongyang'] 값으로 list 인스턴스 b 생성
print(b, type(b))
```

```python
# 시퀀스형 자료 (순서를 가지고 나열된 형태)
c = [1, 3.92, 3 > 5, 'dongyang']
print(c[0], c[1], c[2], c[3])
```

```python
# 시퀀스형 자료 (순서를 가지고 나열된 형태)
# 제어문을 활용하여 한 항목씩 출력하기

d = [1, 3.92, 3 > 5, 'dongyang']
for item in d:
    print(item)
```

```python
# 리스트 변수 값은 수정 가능함 (mutable)
e = [10, 'python', 'java']    # list([10, 'python', 'java'])        # [10, 'python', 'java'] 값으로 list 인스턴스 e 생성
print('변경 전 : ' , e)

e[0] = 'C#'
print('변경 후 : ', e)
```

### 2.5 시퀀스 자료형 - 튜플(tuple)
- 변경이 불가능한(immutable) 원소의 순서 열
- 순서가 있음 : 첨자로 참조할 수 있음


```python
# 튜플 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
a = (10, 20, 30)    # tuple((10, 20, 30))        # (10, 20, 30) 값으로 tuple 인스턴스 a 생성
print(a, type(a))
```

```python
# 튜플에서는 괄호를 생략할 수 있음
b =  10, 20, 30    # 10, 20, 30 값으로 tuple 인스턴스 b 생성
print(b, type(b))
```

```python
# 튜플 자료 항목으로 다양한 자료형 사용 가능
c = (10, 3 > 5, 'computer')   # (10, 3 > 5, 'computer') 값으로 인스턴스 c 생성 (튜플 인스턴스 생성함)
print(c, type(c))
```

```python
# 시퀀스형 자료 (순서를 가지고 나열된 형태)
c = (10, 3 > 5, 'computer')
print(c[0], c[1], c[2])
```

```python
# 시퀀스형 자료 (순서를 가지고 나열된 형태)
# 제어문을 활용하여 한 항목씩 출력하기

c = (10, 3 > 5, 'computer')
for item in c:
    print(item)
```

```python
# 튜플 변수 값은 수정 불가능함 (immutable)
# 튜플은 리스트 자료형 보다 처리 속도 빠름

c = (10, 3 > 5, 'computer')
print('변경 전 : ' , c)

# 변경시 오류 발생
c[0] = 'software'
print('변경 후 : ', c)
```

### 2.6 매핑 자료형 - 딕(dict)
- 키(key)와 값(value)의 매핑이 필요한 경우
- 순서가 없음 : 첨자로 참조할 수 없음, 키로 값을 참조


```python
# dict 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
mart = {'라면':1200, '음료수':1500, '과자':800}    # mart = dict({'라면':1200, '음료수':1500, '과자':800})
print(mart, type(mart))
```

```python
# 키로 값을 참조할 수 있음
mart = {'라면':1200, '음료수':1500, '과자':800}

print(mart['음료수'], mart['과자'])
```

```python
# 키로 값을 변경할 수 있음 (mutable)
mart = {'라면':1200, '음료수':1500, '과자':800}

mart['음료수'] = 5000
print(mart)
```

```python
# 새로운 키와 값을 추가 할 수 있음 (mutable)
mart = {'라면':1200, '음료수':1500, '과자':800}
print('키와 값 추가 전 :', mart)

mart['삼각김밥'] = 1700
print('키와 값 추가 후 :', mart)
```

```python
# 키의 목록, 값의 목록, 각 항목의 목록을 참조할 수 있음
mart = {'라면':1200, '음료수':1500, '과자':800}

print(mart.keys())
print(mart.values())
print(mart.items())
```

```python
# 제어문을 활용하여 키 목록 출력하기
mart = {'라면':1200, '음료수':1500, '과자':800}

for key in mart.keys():
    print(key)
```

```python
# 제어문을 활용하여 값 목록 출력하기
mart = {'라면':1200, '음료수':1500, '과자':800}

for v in mart.values():
    print(v)
```

```python
# 제어문을 활용하여 항목 목록 출력하기
mart = {'라면':1200, '음료수':1500, '과자':800}

for 항목, 값 in mart.items():
    print(항목, 값)
```

```python
# 제어문을 활용하여 항목 목록 출력하기
mart = {'라면':1200, '음료수':1500, '과자':800}

for 항목, 값 in mart.items():
    print(f'{항목} : {값}원')
```

```python
# dict 자료형에서 키는 불변 객체(immutable)만 사용 가능 - 튜플, 숫자, 문자열
mart = {'라면':1200, '음료수':1500, '과자':800}
print('키와 값 추가 전 :', mart)

# 튜플 키 추가 가능
mart[('삼각김밥', '단무지')] = 1700
print('키와 값 추가 후 :', mart)
```

```python
# dict 자료형에서 키는 불변 객체(immutable)만 사용 가능 - 튜플, 숫자, 문자열
mart = {'라면':1200, '음료수':1500, '과자':800}
print('키와 값 추가 전 :', mart)

# 리스트 키 추가 불가능 # 오류 발생
mart[['삼각김밥', '단무지']] = 1700
print('키와 값 추가 후 :', mart)
```

### 2.7 집합형 자료형 - set과 frozenset 
- **set** : 변경이 가능한(mutable) 유일한 원소들의 집합
- **frozenset** : 변경이 불가능한(immutable) 유일한 원소들의 집합
- 순서가 없음 : 첨자로 참조할 수 없음


```python
# set 자료를 변수에 할당하고, 값과 자료형을 출력해 보기
# 중복 저장하지 않음
basket = {'사과', '오렌지', '사과', '배'}    # basket = set({'사과', '오렌지', '사과', '배'})
print(basket, type(basket))
```

```python
# 순서가 없음 : 첨자로 참조할 수 없음
# 오류 발생
basket = {'사과', '오렌지', '배'}    
print(basket[0])
```

```python
# set에 항목 추가
basket = {'사과', '오렌지', '배'}  
print(f'항목 추가 전 : {basket}')

basket.add('딸기')
print(f'항목 추가 후 : {basket}')
```

```python
# set에 항목으로 수정 불가 항목(immutable)인 int, float, str, 튜플은 추가 가능

basket = {'사과', '오렌지', '배'}  
print(f'항목 추가 전 : {basket}')

# set에 수정 불가능 항목 튜플 추가
basket.add(('딸기', '바나나'))
print(f'항목 추가 후 : {basket}')
```

```python
# set에 항목으로 수정 가능 항목(mutable)인 list, dict은 추가  불가능

basket = {'사과', '오렌지', '배'}  
print(f'항목 추가 전 : {basket}')

# set에 수정 가능 항목 리스트 추가
# 오류 발생
basket.add(['딸기', '바나나'])
print(f'항목 추가 후 : {basket}')
```
